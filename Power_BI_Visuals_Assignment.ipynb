# Power BI Visuals Assignment

1. Create a Card Visual to show Total Sales.

- Load Data:
Ensure your global superstore dataset is imported into Power BI.
Select Card Visual: Open the Visualizations pane, click on the Card icon (it looks like a box with "123" inside).
Add Data:
From the Data pane, locate the Sales field. Drag and drop it onto the card visual.
Format Metric:
If the visual displays "First Sales" instead of the total, click the dropdown arrow next to the field name in the visual field well and select Sum.
Customize:
Click on the Format visual (paint brush) icon in the Visualizations pane to customize font size, label formatting, and colors.

2. Create a Donut Chart showing Total Sales by Ship Mode and which ship mode generates the highest sales?

- Load Data: Click Get Data -> Text/CSV and select your Global Superstore dataset.
Add Visualization: In the Visualizations pane, click the Donut Chart icon.
Configure Fields:
Drag the Ship Mode field to the Legend or Category well.
Drag the Sales field to the Values well.
Format: Use the Format pane (brush icon) to adjust data colors, enable labels (category + percentage), and customize the legend for clarity.
Identify Top Ship Mode: The largest segment of the donut visual represents the ship mode with the highest total sales.

Standard class ship mode generates the highest sales.

3. Create a Line Chart showing the Sales Trend over time (by Month and Year). Is there any noticeable seasonal pattern in sales?

- Dominant End-of-Year Dates:
Many of the highest-revenue days shown occurred in November and December (e.g., 18-11-2014, 26-12-2013, 20-11-2014). This could suggest a peak during the holiday season, but without a full chronological dataset to compare against "off-peak" months, this remains an observation rather than a confirmed seasonal trend.
High-Value Outliers:
High sales totals also appear in March (18-03-2011), June (18-06-2014), and September (08-09-2011), indicating that significant revenue spikes occur throughout the year regardless of the season.

4. Create a Bar Chart to compare Sales by Category

- Import Data:
Load your Global Superstore dataset (Excel or CSV) into Power BI Desktop.
Select Visual:
In the Visualizations pane, click on the Clustered Bar Chart or Stacked Bar Chart icon.
Map Fields:
From the Fields pane, drag Category into the Y-Axis (or Axis) area and Sales into the X-Axis (or Values) area.
Format and Refine:
Sort:
Click the three dots () on the top right of the visual, select "Sort axis"
 "Sales", and "Sort descending" to show highest sales first.
Labels:
In the Format pane (), toggle Data labels to On to see exact sales figures.
Title: Rename the title for clarity (e.g., "Total Sales by Category").

7. Create a Scatter Chart to compare Discount vs. Profit by Sub-Category. Are there any outliers
where high discounts resulted in negative profit?

- Select the Visual: In the Visualizations pane, click the Scatter chart icon.
Add Data Fields:
X-Axis:
Drag your Discount field here.
Y-Axis:
Drag your Profit field here.
Values/Details: Drag the Sub-Category field into the Values (or Details) section to display individual data points for each sub-category.
Enhance with Reference Lines: Go to the Analytics pane (magnifying glass icon) and add a Constant Line for the Y-Axis at value 0. This clearly separates positive profit from negative profit.

Negative Profit:
There is only one outlier with a negative profit (located at the bottom left of the chart). However, this point has a low discount value (approximately 250) with sub-category as Tables.
High Discounts:
The data points with the highest discounts (between 600 and 1,100 on the x-axis) all show positive profits (above the 0.0M dashed line) with sub-category as binders.

8. Use a Funnel Chart to show the number of orders by Customer Segment.

- Select the Visual: In the Visualizations pane, click on the Funnel chart icon.
Set the Category: Drag the Segment field into the Category bucket. This creates the bars for each segment.
Set the Values: Drag the Order ID field into the Values bucket.
Tip: Ensure the aggregation is set to Count (Distinct) if you want to count unique orders, or just Count if every row represents one order.

9. Use a Waterfall Chart to analyze how Profit changes by Category within the countries.

- Select the Visual: From the Visualizations pane, click the Waterfall chart icon.
Assign Values (Y-axis): Drag your Profit measure or column into the Y-axis (or Values) field.
Assign Primary Axis (Category): Drag the Category column (e.g., Furniture, Office Supplies, Technology) into the Category (or X-axis) field.

Add the Geographic Breakdown--

Use the Breakdown Field: Drag the Country column into the Breakdown field of the visual.
Impact: This modifies the chart to show how specific countries influence the profit change within each category.
Configuration: By default, Power BI shows the top five influential factors. You can adjust the number of breakdown points in the formatting options to see more or fewer countries.

Refine for Cross-Country Analysis--

Slicers for Filtering: Add a Slicer to your report canvas and drag Country into it. This allows you to select a specific country to see its unique category profit breakdown.
Formatting for Clarity:
Colors: Ensure "Increase" is green and "Decrease" is red to clearly see profitable vs. loss-making categories.
Totals: The final blue bar represents the total profit after all category and country adjustments are accounted for.
Sorting: Use More options (...) > Sort axis to organize the data by Category or Profit value to identify your most impactful segments quickly.

10. Create a Dual-Axis Line Chart to show Total Sales and Profit Over Time (by Month).

- A "Dual-Axis" chart in Power BI is technically called a Line and Stacked Column Chart or a Line and Clustered Column Chart. However, to have two lines, we will use the Line Chart and leverage the secondary Y-axis.
Select the Visual: Click the Line chart icon in the Visualizations pane.
Set the X-Axis: Drag Order Date to the X-axis bucket.
Tip: If it shows a hierarchy (Year, Quarter, Month, Day), click the "X" next to Year, Quarter, and Day to leave only Month.
Set Primary Y-Axis (Sales): Drag your Sales field (or the column containing sales values) to the Y-axis bucket.
Set Secondary Y-Axis (Profit): Drag your Profit field to the Secondary Y-axis bucket.
